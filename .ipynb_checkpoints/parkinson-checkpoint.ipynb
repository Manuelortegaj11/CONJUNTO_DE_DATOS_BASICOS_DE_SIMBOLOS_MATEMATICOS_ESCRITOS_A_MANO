# Parkinson’s Disease Data Set

La enfermedad de Parkinson es una enfermedad progresiva que afecta el sistema nervioso y las partes del cuerpo controladas por los nervios. Los síntomas aparecen lentamente. El primero puede ser un temblor apenas perceptible en una sola mano. Los temblores son comunes, pero este trastorno también puede causar rigidez o disminución del movimiento. [1]

El conjunto de datos consta de varias mediciones biomédicas del habla de 31 personas, 23 de las cuales tienen la enfermedad de Parkinson (EP). Cada columna de la tabla es una medida específica del habla, y cada fila corresponde a una de las 195 grabaciones de estas personas (columna "nombre"). El objetivo principal de los datos es diferenciar entre pacientes sanos y pacientes con EP en función de la columna "Estado", que se establece en 0 para pacientes sanos y 1 para pacientes con EP.

### Reemplazo de valores
Se reemplazan los valores categóricos por valores numéricos float, para que se puedan implementar los métodos de clasificación

La función train_test_split() solo admite valores float, por lo que para poder separar el conjunto de datos en subconjuntos de entrenamiento y prueba, se hace necesario realizar los siguientes reemplazos

````python
# se hace una copia del conjunto de datos
data = df.copy()

# se remueve la columna name ya que no es importante para el entrenamiento y la clasificacion
feature = data.drop(["name"],axis=1)
answer = data['status'].values.reshape(-1)
````

````python
# se reemplazan los varoles no numéricos por Dummies tipo float
le = LabelEncoder()
for column_name in feature.columns:
  if feature[column_name].dtype == object:
    feature[column_name] = le.fit_transform(feature[column_name])
  else:
    pass

answer = le.fit_transform(answer)
````

````python
feature.head()
````

Se establecen los conjuntos de entrenamiento y pruebas

````python
xtrain,xtest,ytrain,ytest = train_test_split(feature,answer,test_size=0.2,random_state=42)
````

# 5. Experimento


Para el experimento se llevará a cabo una prueba inicial donde se medirá el rendimiento de cada algoritmo teniendo en cuenta la precicion y tiempo de demora en alcanzar el resultado, se evaluarán 3 modelos de calisificación y se realizará una comparación entre los mismos

Modelos:
*   RandomForest
*   DecisionTree
*   Neural network using Multilayer perceptron

## **Random Forest**

Prueba con valores iniciales del algoritmo **RandomForestClassifier**

````python
clf = RandomForestClassifier()
clf.fit(xtrain,ytrain)
y_pred = clf.predict(xtest)
y_prob = clf.predict_proba(xtest)

mapping = dict(zip(le.classes_, range(0, len(le.classes_)+1)))
print(mapping)

cm = confusion_matrix(ytest,y_pred)
cm_df = pd.DataFrame(cm,index=mapping)

rf_precision = 100 * accuracy_score(ytest,y_pred) 
print('Precisión: {0:.3f}%'.format(rf_precision))

plt.figure(figsize=(8,6))                  
sns.heatmap(cm_df, annot=True)
plt.title('RandomForestClassifier (initial)')
plt.ylabel('Valor real')
plt.xlabel('Valor estimado')
plt.show()
````

Para la prueba inicial se obtuvo un puntaje de 97.436% con un tiempo de 0.776 segundos

Utilizando GridSeachCV para RandomForestClassifier

````python
param_grid = {'n_estimators':[80,90,100,110,120],
              'criterion':['gini','entropy'],
              'max_depth':[1,3,5,9],
              }

grid = GridSearchCV(RandomForestClassifier(random_state=42),param_grid,cv=5)
grid.fit(xtrain,ytrain)
print(grid.best_params_)
````

Mejores parametros
```
  best_params = {'criterion': 'gini', 'max_depth': 5, 'n_estimators': 80}
```

Prueba con mejor combinación de parámetros para el algoritmo **RandomForestClassifier**

````python
clf = RandomForestClassifier(criterion='gini',max_depth=5,n_estimators=80 )
clf.fit(xtrain,ytrain)
y_pred = clf.predict(xtest)
y_prob = clf.predict_proba(xtest)

mapping = dict(zip(le.classes_, range(0, len(le.classes_)+1)))
print(mapping)

cm = confusion_matrix(ytest,y_pred)
cm_df = pd.DataFrame(cm,index=mapping)

rf_precision = 100 * accuracy_score(ytest,y_pred) 
print('Precisión: {0:.3f}%'.format(rf_precision))

plt.figure(figsize=(8,6))                  
sns.heatmap(cm_df, annot=True)
plt.title('RandomForestClassifier (best params)')
plt.ylabel('Valor real')
plt.xlabel('Valor estimado')
plt.show()
````

La prueba con la mejor conbinación de parámetros

```
  criterion: gini
  max_depth: 5
  n_estimators: 80
```
dio como resultado 100% de precición y el modelo converge

## **Decision Tree**

Prueba con valores iniciales del algoritmo **DecisionTreeClassifier**

````python
clf = DecisionTreeClassifier()
clf.fit(xtrain,ytrain)
y_pred = clf.predict(xtest)
y_prob = clf.predict_proba(xtest)

mapping = dict(zip(le.classes_, range(0, len(le.classes_)+1)))

cm = confusion_matrix(ytest,y_pred)
cm_df = pd.DataFrame(cm,index=mapping)

dt_precision = 100 * accuracy_score(ytest,y_pred) 
print('Precisión: {0:.3f}%'.format(dt_precision))

plt.figure(figsize=(8,6))                  
sns.heatmap(cm_df, annot=True)
plt.title('DecisionTreeClassifier (initial)')
plt.ylabel('Valor real')
plt.xlabel('Valor estimado')
plt.show()
````

Para la prueba inicial se registró una precisión de 92.308%, esta prueba se realizó con los valores por defecto del algortimo

Utilizando GridSeachCV para DecisionTreeClassifier

````python
param_grid = {'criterion':['gini', 'entropy'],
              'splitter':['best','random'],
              'max_depth':[1,3,5,9],
              }

grid = GridSearchCV(DecisionTreeClassifier(random_state=42),param_grid,cv=5)
grid.fit(xtrain,ytrain)
print(grid.best_params_)
````

Mejores parametros 
```
  best_params = {'criterion': 'gini', 'max_depth': 1, 'splitter': 'best'}
```

````python
clf = DecisionTreeClassifier(criterion='gini',max_depth=1,splitter='best')
clf.fit(xtrain,ytrain)
y_pred = clf.predict(xtest)
y_prob = clf.predict_proba(xtest)

mapping = dict(zip(le.classes_, range(0, len(le.classes_)+1)))

cm = confusion_matrix(ytest,y_pred)
cm_df = pd.DataFrame(cm,index=mapping)

dt_precision = 100 * accuracy_score(ytest,y_pred) 
print('Precisión: {0:.3f}%'.format(dt_precision))

plt.figure(figsize=(8,6))                  
sns.heatmap(cm_df, annot=True)
plt.title('DecisionTreeClassifier (best params)')
plt.ylabel('Valor real')
plt.xlabel('Valor estimado')
plt.show()
````

En la prueba con mejores parametros se obtuvo que la precisión fue del 100%, logrando clasificar correctamente todos los datos del conjuto de pruebas

GridSearch para MLP

````python
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.fit_transform(xtest)

param_grid = {
    'hidden_layer_sizes': [(8,4,6),(16,8,12),(24,20,16),(32,22,26)],
    'alpha':[0.001, 0.01, 0.1],
    'max_iter': [150, 300, 350, 400, 450],
    'solver': ['lbfgs', 'sgd', 'adam']
}

grid = GridSearchCV(MLPClassifier(random_state=42),param_grid=param_grid,cv=5)
grid.fit(xtrain,ytrain)
print(grid.best_params_)
````

Mejores parametros 

```
best_params = {'alpha': 0.001, 'hidden_layer_sizes': (8, 4, 6), 'max_iter': 150, 'solver': 'lbfgs'}
```



````python
clf = MLPClassifier(alpha=0.001, hidden_layer_sizes=(8, 4, 6), max_iter=150, solver='lbfgs')
clf.fit(xtrain,ytrain)

y_pred = clf.predict(xtest)
y_prob = clf.predict_proba(xtest)

mapping = dict(zip(le.classes_, range(0, len(le.classes_)+1)))
print(mapping)

cm = confusion_matrix(ytest,y_pred)
cm_df = pd.DataFrame(cm,index=mapping)

mlp_precision = 100 * accuracy_score(ytest,y_pred) 
print('Precisión: {0:.3f}%'.format(mlp_precision))

plt.figure(figsize=(8,6))                  
sns.heatmap(cm_df, annot=True)
plt.title('MLP (best params)')
plt.ylabel('Valor real')
plt.xlabel('Valor estimado')
plt.show()
````

Teniendo en cuenta el warning:
```
  lbfgs failed to converge (status=1):
  STOP: TOTAL NO. of ITERATIONS REACHED LIMIT.
```
procedemos a aumentar el número máximo de iteraciones con el fin de permitirle al modelo converger y llegar a un puntaje de precisión más alto

el nuevo limite de iteraciones se establece arbitrariamente en:
```
  max_iter: 300
```

Se ejecuta nuevamente la mejor conbinación de parámetros con el ajuste a   ```max_iter``` y se revisan los resultados

````python
clf = MLPClassifier(alpha=0.001, hidden_layer_sizes=(8, 4, 6), max_iter=300, solver='lbfgs')
clf.fit(xtrain,ytrain)

y_pred = clf.predict(xtest)
y_prob = clf.predict_proba(xtest)

mapping = dict(zip(le.classes_, range(0, len(le.classes_)+1)))
print(mapping)

cm = confusion_matrix(ytest,y_pred)
cm_df = pd.DataFrame(cm,index=mapping)

mlp_precision = 100 * accuracy_score(ytest,y_pred) 
print('Precisión: {0:.3f}%'.format(mlp_precision))

plt.figure(figsize=(8,6))                  
sns.heatmap(cm_df, annot=True)
plt.title('MLP (best params)')
plt.ylabel('Valor real')
plt.xlabel('Valor estimado')
plt.show()
````

Finalmente obtenemos como resultado un 100% de precisión para el modelo MLP
